## Filtering data to explore majors in colleges with the highest Median Earnings

step 1: keeping columns which names contain 'major', 'institution', 'earnings'

```python
majors= data.filter(regex='major|institution|earnings')
majors.head()
del majors['earnings_more_than_25k']

majors.head()
```

step 2: removing NaN for 'median_earnings' columns

```python
print(majors['median_earnings'])
```

```python
majors['median_earnings'].dropna()
```

step 3: filtering data to keep only colleges with Median Earnings above 80000

```python
majors = majors[majors['median_earnings'] > 80000]
majors
```

```python
# replacing NaN with 0
majors = majors.fillna(0) 
majors
```

```python
# setting index to institution_name
majors.set_index("institution_name",inplace=True)
majors.head()
```

```python
print(majors.columns.values)
```

```python
del majors['median_earnings']
```

```python
majors.head()
```

```python
# finding average percentage for each major
majors.loc['mean_score'] = majors.mean()
majors
```

```python
# transpose dataframe
majors = majors.loc['mean_score']
majors
```

```python
# setting name for the majors column
majors.index.name = 'majors'
majors
```

```python
majors.to_csv('C:\\Users\\chiku\\OneDrive\\Documents\\Gingko Assessment\\majors.csv')
```
